```python
##Transformando colunas referente ao tempo de expulsão em uma só
dataset['minuto_expulsao_um'] = False
dataset['minuto_expulsao_dois'] = False
dataset['minuto_expulsao_tres'] = False
dataset['minuto_expulsao_quatro'] = False
dataset['minuto_expulsao_cinco'] = False
dataset['minuto_expulsao_seis'] = False
dataset['minuto_expulsao_sete'] = False
dataset['minuto_expulsao_oito'] = False
dataset['qtd_expulsoes'] = False

dataset['diferenca_gols_expulsao_um'] = False
dataset['diferenca_gols_expulsao_dois'] = False

for index, row in dataset.iterrows():
    
    dataset.loc[index,'diferenca_gols_expulsao_um'] = dataset.loc[index,'quantidade_de_gols_no_momento_expulsao_um_time_a'] - dataset.loc[index,'quantidade_de_gols_no_momento_expulsao_um_time_b']
    dataset.loc[index,'diferenca_gols_expulsao_dois'] = dataset.loc[index,'quantidade_de_gols_no_momento_expulsao_dois_time_a'] - dataset.loc[index,'quantidade_de_gols_no_momento_expulsao_dois_time_b']
    ##Expulsões
    minutos_expulsoes = []
    expulsao = False
    if(dataset.loc[index,'minuto_primeira_expulsao_time_a']):
        minutos_expulsoes.append(dataset.loc[index,'minuto_primeira_expulsao_time_a'])
        expulsao = True
    if(dataset.loc[index,'minuto_primeira_expulsao_time_b']):
        minutos_expulsoes.append(dataset.loc[index,'minuto_primeira_expulsao_time_b'])
        expulsao = True
        
    if(expulsao):
        expulsao = False
        if(dataset.loc[index,'minuto_segunda_expulsao_time_a']):
            minutos_expulsoes.append(dataset.loc[index,'minuto_segunda_expulsao_time_a'])
            expulsao = True
        if(dataset.loc[index,'minuto_segunda_expulsao_time_b']):
            minutos_expulsoes.append(dataset.loc[index,'minuto_segunda_expulsao_time_b'])
            expulsao = True

        if(expulsao):
            expulsao = False
            if(dataset.loc[index,'minuto_terceira_expulsao_time_a']):
                minutos_expulsoes.append(dataset.loc[index,'minuto_terceira_expulsao_time_a'])
                expulsao = True
            if(dataset.loc[index,'minuto_terceira_expulsao_time_b']):
                minutos_expulsoes.append(dataset.loc[index,'minuto_terceira_expulsao_time_b'])
                expulsao = True

            if(expulsao):
                if(dataset.loc[index,'minuto_quarta_expulsao_time_a']):
                    minutos_expulsoes.append(dataset.loc[index,'minuto_quarta_expulsao_time_a'])
                if(dataset.loc[index,'minuto_quarta_expulsao_time_b']):
                    minutos_expulsoes.append(dataset.loc[index,'minuto_quarta_expulsao_time_b'])
        minutos_expulsoes.sort()
        
    dataset.loc[index,'qtd_expulsoes'] = len(minutos_expulsoes)
    
    i = 0
    while(i < len(minutos_expulsoes)):
        if(i == 0):
            dataset.loc[index,'minuto_expulsao_um'] = minutos_expulsoes[i]
        elif(i == 1):
             dataset.loc[index,'minuto_expulsao_dois'] = minutos_expulsoes[i]
        elif(i == 2):
             dataset.loc[index,'minuto_expulsao_tres'] = minutos_expulsoes[i]
        elif(i == 3):
             dataset.loc[index,'minuto_expulsao_quatro'] = minutos_expulsoes[i]
        elif(i == 4):
             dataset.loc[index,'minuto_expulsao_cinco'] = minutos_expulsoes[i]
        elif(i == 5):
             dataset.loc[index,'minuto_expulsao_seis'] = minutos_expulsoes[i]
        elif(i == 6):
             dataset.loc[index,'minuto_expulsao_sete'] = minutos_expulsoes[i]
        elif(i == 7):
             dataset.loc[index,'minuto_expulsao_oito'] = minutos_expulsoes[i]
        i += 1
    
```
